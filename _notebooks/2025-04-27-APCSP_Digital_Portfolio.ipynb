```python
import random

# List of quiz questions, each stored as a dictionary
quiz_questions = [
    {"question": "What is the molar mass of KCl?", "options": ["74.55", "39.10", "58.44", "18.02"], "answer": "74.55"},
    {"question": "What is the value of Avogadro's constant?", "options": ["6.022e23", "3.14", "9.81", "8.314"], "answer": "6.022e23"},
    {"question": "Which gas law is PV=nRT?", "options": ["Ideal Gas Law", "Boyle's Law", "Charles's Law", "Henry's Law"], "answer": "Ideal Gas Law"},
    {"question": "What is the pH of a neutral solution at 25°C?", "options": ["7", "0", "14", "1"], "answer": "7"},
    {"question": "Which ion has a charge of +1?", "options": ["Na+", "Cl-", "O2-", "Mg2+"], "answer": "Na+"}
]

# Randomize question order for variety
random.shuffle(quiz_questions)

# Student-developed procedure with:
# - Name: ask
# - Parameter: q (the question dictionary)
# - Return: Boolean (True if answer is correct, False otherwise)
# - Contains sequencing, selection, and iteration
def ask(q):
    print(f"\n{q['question']}")
    for i, opt in enumerate(q['options']):
        print(f"{chr(65+i)}. {opt}")
    
    # Input from user
    ans = input("Answer (A-D): ").strip().upper()
    while ans not in ['A', 'B', 'C', 'D']:
        ans = input("Try again (A-D): ").strip().upper()
    
    # Selection: check if the answer is correct
    return q['options'][ord(ans) - 65] == q['answer']

# Main function to run the game
def start():
    name = input("Enter your name: ")  # Input from user
    print(f"\nGood luck, {name}!")

    score = 0
    # Iteration over questions
    for q in quiz_questions:
        if ask(q):
            print("Correct!")     # Output
            score += 1
        else:
            print("Incorrect.")   # Output

    # Final output
    print(f"\n{name}, your score: {score}/{len(quiz_questions)}")

# Start the game
start()





#  GPT Wrote the loop & selection part of the code
```
